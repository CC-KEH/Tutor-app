```python
def store_chunks_in_pinecone(chunks,index,namespace):
    for idx, chunk in enumerate(chunks):
        text = chunk.page_content
        text = text.replace("\n",". ")
        text = text.replace("\t",". ")

        vector = encoder.encode(text)
        vector_id = f"doc_chunk_{idx}"
        
        index.upsert(
            vectors = 
            [
                {
                    "id" : vector_id,
                    "values" : vector,
                    "metadata" : {'source': '','text' : text}
                }
            ] ,
            namespace=namespace,
        )
```
